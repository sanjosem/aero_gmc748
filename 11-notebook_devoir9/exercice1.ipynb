```python
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'retina'
```

Le coefficient ci-dessous permet de modifier la valeur de l'énoncé (erronée) pour vérifier les calculs et les tracés. C'est une variable externe utilisé dans les fonctions définies par la suite. Le modifier et rappeler les fonctions pour les évaluations. 

```python
coeff_enonce = 0.086
```

```python
def Theta_lam(ReL,X):
    theta_L = 0.664* ReL**(-1/2) * X**(1/2)
    return theta_L
```

```python
def Theta_turb(ReL,XT,X):
    theta_L = (0.6118*ReL**(-3/5)*XT**(3/5) +6/5*coeff_enonce*ReL**(-1/5)*(X-XT))**(5/6)
    return theta_L
```

```python
def Cf_lam(ReL,X):
    Cf = 0.664* ReL**(-1/2) * X**(-1/2)
    return Cf
```

```python
def Cf_turb(ReL,XT,X):
    ReT=ReL*XT
    Rex=ReL*X
    c1 = 0.6118
    c2 = coeff_enonce * 6/5
    Cf = 2*coeff_enonce*(c1*ReT**(3/5) +c2*(Rex-ReT))**(-1/6)
    return Cf
```

```python
def CD(ReL,XT):
    ReT=ReL*XT
    CD = 4 * Theta_turb(ReL,XT,1)
    return CD
```

### Question 3 

```python
XT=np.linspace(0,1)
ReL=10**7
plt.plot(XT,CD(ReL,XT));
plt.xlabel(r'$X_T$');
plt.ylabel(r'$C_{D,v}$');
plt.tight_layout();
plt.savefig('Coeff_trainee_{0:.0f}.png'.format(coeff_enonce/0.0086),dpi=100)
```

### Question 4

```python
XT=0.2
ReL=10**7
X=np.linspace(0.,1.,100)
Xlam=X[X<=XT]
Xturb=X[X>XT]
plt.plot(Xlam,Theta_lam(ReL,Xlam));
plt.plot(Xturb,Theta_turb(ReL,XT,Xturb));
plt.xlabel(r'$X$');
plt.ylabel(r'$\Theta/L$');
plt.tight_layout();
plt.savefig('Theta_evol_{0:.0f}.png'.format(coeff_enonce/0.0086),dpi=100)
```

```python
plt.plot(Xlam,Cf_lam(ReL,Xlam));
plt.plot(Xturb,Cf_turb(ReL,XT,Xturb));
plt.xlabel(r'$X$');
plt.ylabel(r'$C_f$');
plt.tight_layout();
plt.savefig('Coeff_frottement_{0:.0f}.png'.format(coeff_enonce/0.0086),dpi=100)
```
